   The following coordinate map is a function that takes a real number $t$ (usually representing time) as input and maps it to $A$, that represents the amplitude of the audio signal at that specific time point.

   $$
   \text{Coordinate Map}: \mathbb{R} \to \mathbb{R} : [t] \to [A]
   $$

   Here, $t$ belongs to the set of real numbers $\mathbb{R}$, and it is mapped to the amplitude value $A$ at time $t$.

   The following map is a feature map, a higher dimensional representation, which processes a collection of features obtained by RFF to compute a corresponding the value $A$, which is the reconstructed amplitude of the audio signal.

   $$
   \text{Feature Map}: \mathbb{R}^{\text{n}} \to \mathbb{R} : [t_{1 \cdots n}] \to [A]
   $$

   This final map combines the two previous mappings to describe the entire transformation used for audio reconstruction.
   
   It first maps a single time point $t$ to a set of features $t_{1 \cdots n}$ and then applies the feature map to obtain the final amplitude value $A$.

   $$
   \text{Final Map}: \mathbb{R} \to \mathbb{R}^{\text{features}} \to \mathbb{R} : [t] \to [t_{1 \cdots k}] \to [A]
   $$

   This final mapping shows the complete process where an input time point $t$ is first transformed into a feature space (of higher dimension) and then mapped to the corresponding reconstructed amplitude value $A$.

In audio reconstruction, we aim to reconstruct the amplitude values $A$ of an audio signal from its sampled data points over time. The coordinate map captures the relationship between the individual time points and their corresponding amplitude values. The feature map extends this by considering derived features to produce a higher dimensional representation. Finally, the final map illustrates the full transformation where the original time-domain information is transformed into a feature space and then mapped to reconstruct the audio signal accurately.

We are normalizing in the range of -100 to 100 instead of -1 to 1 because of several reasons:
1. Using a larger range like -100 to 100 can make the **differences between time points more pronounced**. When the time values are spread out over a larger range, **the RFF transformation can capture more distinct variations in the data**. In audio reconstruction, the **subtle differences in amplitude over time need to be accurately modeled**. A smaller range like -1 to 1 might compress these variations too much, making it harder for the model to distinguish small changes in the input space. The larger range can help in representing the audio signal more effectively.
2. It provides a good spread of values, which can help the RFF to capture patterns at different scales. The RFF method relies on random projections to approximate kernel functions, and **having a wider range of input values can help in capturing a diverse set of patterns** in the data. By using a larger range, we can ensure that the RFF features can effectively capture the underlying structure of the audio signal. It's large enough to provide good resolution, but not so large as to potentially cause numerical instability issues.
3. Choice based on results--the choice of the range was based on the results of experimentation. Using -1 to 1 produced worse results compared to -100 to 100. 

```python
# here we are creating a tensor of shape (num_samples, 1) where each value is the time in seconds
# creating it from 0 to len(audio) and then normalizing it to be between -100 and 100
X = torch.arange(0, len(audio)).unsqueeze(1).float()
X = X / X.max() * 200 - 100
print(X.shape)
print(X)
```

```python
Audio(audio, rate=sampling_rate)
```

```python
def create_rff_features(X, num_features, sigma):
    # set a random state to the sampler for repoducibility
    rff = RBFSampler(n_components=num_features, gamma=1 /
                     (2 * sigma**2), random_state=13)
    X = X.cpu().numpy()
    X = rff.fit_transform(X)
    return torch.tensor(X, dtype=torch.float32)
```

```python
num_features = 20000 # number of features to select for RFF
sigma = 0.01 # the parameter of RBF sampler

X_rff = create_rff_features(X, num_features, sigma)
print(X_rff.shape)
print(X_rff)
```

```python
audio.shape
```

```python
# fit the linear regression model
model = LinearRegression()
model.fit(X_rff.numpy(), audio)

pred_audio = model.predict(X_rff.numpy())
print(pred_audio.shape)
```

```python
sf.write('pred_audio.wav', pred_audio, sampling_rate) # update the path to save the file in the current directory
```

```python
print(pred_audio)
print(audio)
```

```python
Audio(pred_audio, rate=sampling_rate)
```

```python
plot_audio(pred_audio, sampling_rate, 'red', 'Reconstructed Audio Waveform')
```

```python
# compare the result
plt.figure(figsize=(15, 4))
plt.plot(audio, color = 'black', alpha = 1, label = 'Original Audio')
plt.plot(pred_audio, color = 'white', alpha = 0.9, label = 'Reconstructed Audio')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Original vs Reconstructed Audio Waveform')
plt.grid()
plt.legend()
plt.show()
```

```python
def calculate_rmse_snr(original, reconstructed, max_value):
  mse = np.mean(np.square(original - reconstructed).numpy())
  rmse = np.sqrt(mse)
  snr = 10 * np.log10( (max_value**2) / mse )
  return rmse, snr
```

```python
rmse, snr = calculate_rmse_snr(audio, pred_audio, pred_audio.max())
print("RMSE: ", rmse)
print("SNR: ", snr)
```
